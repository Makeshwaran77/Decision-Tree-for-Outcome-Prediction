```python
import matplotlib.pyplot as plt

top_five_items_qty = d.groupby('Item Name')['Qty.'].sum().nlargest(5)

plt.figure(figsize=(7, 6))
top_five_items_qty.plot(kind='bar', color='black')
plt.title('Top 5 Items Sold by Quantity')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity Sold')
plt.show()
```
